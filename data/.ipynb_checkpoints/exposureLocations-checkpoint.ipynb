```python
sa2 = pd.read_csv("sa2.csv")
geoSa2 = gpd.read_file("geoSa2.geojson")
geoSa2.plot(figsize=(12, 12))
geoSa2
```

```python
import numpy as np

resultsSa2 = gpd.GeoDataFrame(columns=["name", "time", "pop", "geometry"])

for i in data.index:
    geodatum = geoSa2[geoSa2["geometry"].contains(data.at[i, "geometry"])]
    code = geodatum.iloc[0]["sa2Code"]
    datum = sa2[sa2["sa2Code"] == int(code)]
    if code in resultsSa2.index:
        resultsSa2.at[code, "time"] = data.at[i, "time"] + resultsSa2.at[code, "time"]
    else:
        resultsSa2.at[code, "time"] = data.at[i, "time"]
        resultsSa2.at[code, "pop"] = datum.iloc[0]["value"]
        resultsSa2.at[code, "name"] = geodatum.iloc[0]["sa2Name"]
        resultsSa2.at[code, "geometry"] = geodatum.iloc[0]["geometry"]

for i in geoSa2.index:
    if geoSa2.at[i, "sa2Code"] not in resultsSa2.index:
        resultsSa2.at[geoSa2.at[i, "sa2Code"], "name"] = geoSa2.at[i, "sa2Name"]
        resultsSa2.at[geoSa2.at[i, "sa2Code"], "geometry"] = geoSa2.at[i, "geometry"]
        resultsSa2.at[geoSa2.at[i, "sa2Code"], "time"] = 0
        resultsSa2.at[geoSa2.at[i, "sa2Code"], "pop"] = sa2[sa2["sa2Code"] == int(geoSa2.at[i, "sa2Code"])].iat[0, 2] if len(sa2[sa2["sa2Code"] == int(geoSa2.at[i, "sa2Code"])]) == 1 else 0

resultsSa2.index.name = "code"
resultsSa2 = resultsSa2.reset_index()
resultsSa2["code"] = pd.to_numeric(resultsSa2["code"])
resultsSa2["time"] = pd.to_numeric(resultsSa2["time"])
resultsSa2["pop"] = pd.to_numeric(resultsSa2["pop"])
resultsSa2.to_file("resultsSa2.geojson", driver="GeoJSON")
resultsSa2
```

```python
fig, ax = plt.subplots(
    tight_layout=True,
    figsize=(12, 8)
)
resultsSa2.query("pop > 0").plot.scatter(x="pop", y="time", ax=ax)
```

```python
resultsSa2["exposure"] = resultsSa2["time"] / resultsSa2["pop"]
resultsSa2.query("(pop > 750) & (name == 'Hackett')").sort_values("exposure").head(60)
```
